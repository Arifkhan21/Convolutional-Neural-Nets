```python
import numpy as np 
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
im = cv.imread('hacktoberfest.png')
img=cv.cvtColor(im,cv.COLOR_RGB2GRAY)
plt.imshow(img,cmap='gray')
```

```python
kernel = [[-1.0,-1.0,-1.0],
          [-1.0,8.0,-1.0],
         [-1.0,-1.0,-1.0]]
kernel = np.array(kernel)
```

```python
new_img = np.zeros((img.shape[0]-kernel.shape[0]+1,img.shape[1]-kernel.shape[1]+1))

print(img.shape)
print(new_img.shape)
```

```python
for ix in range(new_img.shape[0]):
    for jx in range(new_img.shape[1]):
        
        patch = img[ix:ix+kernel.shape[0],jx:jx+kernel.shape[1]]
        val = patch*kernel
        
        new_img[ix,jx]=max(0,val.sum())
```

```python
plt.imshow(new_img,cmap='gray')
```
